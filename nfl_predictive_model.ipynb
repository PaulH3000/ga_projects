```python
#import xml.etree.ElementTree as ET
import lxml.etree as ET #more powerful xml library
import pandas as pd

#TODO: need to prefix the attribute names with the tag name of enclosing element
# (when populating the dictionary keys)
# note this returns df with singel record = total stat's over all post season games in 2017 for the Saints

#class encapsulating methods to convert XML data into a dataframe
class XML2DataFrame:

    def __init__(self, xml_data):
        self.root = ET.XML(xml_data)

    def parse_root(self, root):
        """Return a dictionary populated from the data and
         attributes of the children nodes under this XML root."""
        return [self.parse_element(child) for child in iter(root)]

    def parse_element(self, element, parsed=None):
        """ Populate dictionary with {attrib_name:attrib_value} and {tag:text} from
         this XML element and all its children."""
        if parsed is None:
            parsed = dict()
        for key in element.keys():
            parsed[key] = element.attrib.get(key)
        if element.text:
            parsed[element.tag] = element.text
        
        """ Apply recursion"""
        for child in list(element):
            self.parse_element(child, parsed)
        return parsed

    def process_data(self):
        structure_data = self.parse_root(self.root)
        return pd.DataFrame(structure_data)
```
